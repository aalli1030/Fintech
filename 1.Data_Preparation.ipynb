## Format conversion of feature names 

```python
for i in range (len(current_fea)):
    current_fea[i] = str.lower(str(current_fea[i]))
    current_fea[i] = str(current_fea[i]).replace('_','')
```

```python
current_fea
```

```python
for i in range (len(previous_fea)):
    previous_fea[i] = str(previous_fea[i]).replace('_','')
```

```python
previous_fea
```

```python
common_fea = set(previous_fea)&(set(current_fea))
print (len(common_fea))
common_fea
```

### Convert column names in the data frame and save dataframe with common features and issued, loanstatus

```python
# The last two rows of each file Q1~Q4 are NAN.
df_2017=pd.concat([df_2017_Q1[:-2], df_2017_Q2[:-2], df_2017_Q3[:-2], df_2017_Q4[:-2]])
```

```python
df_2017.shape
```

```python
df_2017.isnull().sum()
```

```python
for col in df_2017.columns.values:
    df_2017.rename(columns={str(col):str(col).replace('_','')}, inplace=True)
```

```python
df_2017new=df_2017[list(common_fea)+['issued', 'loanstatus']]
```
